```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
def load_and_process(url_or_path_to_csv_file):
    # Method Chain 1 (Load data, deal with missing data, and making data readable)
    df1 = (
        pd.read_csv(url_or_path_to_csv_file)
        .dropna()
        .rename(columns ={"LOCAL_DATE":"DATE"})
        .reset_index(drop = True)
        .sort_values("DATE", ascending=False)
      )
    #Method Chain 2 (Creating new columns for different regions of Canada)
    df2 = (
        df1
        .assign(MEAN_TEMPERATURE_ATLANTIC = (df1['MEAN_TEMPERATURE_STJOHNS'] + df1['MEAN_TEMPERATURE_MONCTON'] +df1['MEAN_TEMPERATURE_HALIFAX'])/3)
        .drop
        .rename(columns ={"MEAN_TEMPERATURE_TORONTO":"MEAN_TEMPERATURE_ONTARIO"})
        .rename(columns ={"MEAN_TEMPERATURE_MONTREAL":"MEAN_TEMPERATURE_QUEBEC"})
    )
        
    return df2
```

```python
load_and_process('Canadian_climate_history (1970-2020).csv')
```
